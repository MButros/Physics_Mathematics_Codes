Find the acceleration of the blocks and the tension in the string.

Initialize some variables:

```python
g = 9.8 # m/s^2
M = 3.3 # kg
m = 2.1 # kg
```

From the free body diagram the sum of the forces on the hanging mass (m):
$$ T - mg = -ma$$ 
the sum of forces in the x direction on the sliding mass (M):
$$ T = Ma $$
Solving for $a$ and $T$ we get:

```python
a = (m*g)/(M+m)
T = m*(g-a)
print('The acceleration of both blocks is ' '%0.3f' % a, 'meters per second squared')
print('The tension in the string is ' '%0.3f' % T, 'newtons')
```
