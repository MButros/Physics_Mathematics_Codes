Using Euler's method to solve the differential equation $$ y'= y + \sin(x) \textrm{ with } y(0) = 1$$

Euler's method is what you will need to solve the computational project, keep in mind that the differential equation you will be solving for the project will be different.


```python
import numpy as np
from matplotlib import pyplot as plt
```

Initialize some values

```python
x0 = 0
y0 = 1
xf = 10
n = 101
```

Setting up Euler's method on the given differential equation

```python
delta_x = (xf-x0)/(n-1)

x = np.linspace(x0,xf,n)
y = np.zeros([n])

y[0] = y0
```

Implementing Euler's method and plotting the results

```python
for i in range(1, n):
	y[i] = delta_x*(-y[i-1]+np.sin(x[i-1]))+y[i-1]

for i in range(n):
	print(x[i], y[i])

plt.plot(x, y, 'o')
plt.xlabel('Value of x')
plt.ylabel('Value of y')
plt.title('Approximate Solution with Forward Euler’s Method')
plt.show()
```
