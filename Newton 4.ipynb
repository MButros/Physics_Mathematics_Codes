Example of Newton's law with friction

Initialoze some Variables

```python
import math 
import numpy as np
g = 9.8 # in m/s^2
m = 75 # in kg
mu = 0.1 # coefficient of friction (kinetic)
phi = 42*np.pi/180 # in rad
```

Sum of the forces in x and y directiosn give
$$ T\cos(phi) - \mu N = 0$$
$$ T\sin(phi) + N = mg $$
Which means we can set up the equations in matrices as

```python
A = np.array([[np.cos(phi), -mu],[np.sin(phi), 1]])
b = np.array([0, m*g])
```

Now we solve the system 

```python
A_inv = np.linalg.inv(A)
```

```python
F = A_inv * b 
```

```python
print('The tension in the string is ' '%0.3f ' % F[0,1], 'newtons')
```

```python
print('The normal force is ' '%0.3f ' % F[1,1], 'newtons')
```
