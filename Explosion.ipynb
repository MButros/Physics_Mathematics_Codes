Example of finding the velocity of a part from an explosion 

```python
import numpy as np
import math
```

```python
M = 1 # total percentage of original mass
M_c = 0.3*M # relative mass of piece C
V_c = 5 # in m/s (in two directions)
M_b = 0.2*M
# This means that M_a = M -(M_b+M_c) = 0.5M
M_a = 0.5*M
```

Conservation of linear momentum in y direction
$$ P_{fA,y} = 0 $$
$$ P_{fB,y} = -0.2MV_{fb,y} = -0.2MV_{fb}\sin(50)$$
$$ P_{fC,y} = 0.30MV_{fc,y} = 0.3MV_{fc}\sin(80)$$

The last equation allows us to solve for $V_{fb}$

```python
V_fb = (0.3*M*5*np.sin(80*np.pi/180))/(0.2*M*np.sin(50*np.pi/180)) 
print('The final velocity of piece B is ' '%0.3f' %V_fb, ' meters per second')
```

Conservation of linear momentum in y direction
$$ P_{fA,x} = 0.5MV_{fa,x} = 0.5MV_{fa} $$
$$ P_{fB,x} = -0.2MV_{fb,x} = 0.2MV_{fb}\cos(50)$$
$$ P_{fC,x} = 0.30MV_{fc,x} = 0.3MV_{fc}\cos(80)$$

This allows us to solve for $V_{fa}$

```python
V_fa = (0.2*M*V_fb*np.cos(50*np.pi/180)+0.3*M*V_c*np.cos(80*np.pi/180))/(0.5*M)
print('The final velocity of piece A is ' '%0.3f' %V_fa, ' meters per second')
```
