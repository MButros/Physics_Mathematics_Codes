Find the tension in the cords

Initiaizee some variables

```python
g = 9.8 # m/s^2
m = 15  # kg 
```

Sum of the forces on the mass 

```python
T = m*g 
```

Applying Newton's second law of motion to the cords $T_1$ and $T_2$ we get
$$T_1\cos(28) - T_2\cos(47) = 0$$
$$T_1\sin(28) +T_2\sin(47) = T $$
Setting the system above in two matrices $A$ for the left hand side coefficients of $T_1$ and $T_2$ and $b$ for the right hand side values we get


```python
import numpy as np
A = np.array([[np.cos(28*np.pi/180), -np.cos(47*np.pi/180)], [np.sin(28*np.pi/180), np.sin(47*np.pi/180)]])
b = np.array([0, T])
```

The values of the cords are given by the solution matrix given by the matrix $A^{-1}b$

```python
Sol = np.linalg.inv(A)*b
```

```python
print('The value of T_1 is ' '%0.3f' % Sol[0,1], 'newtons')
print('The value of T_2 is ' '%0.3f' % Sol[1,1], 'newtons')
print('The value of T is ' '%0.3f' % T, 'newtons' )
```
